###An average MNIST digit looks like a digitally hand-written digit

######Compute average digits, display based on these averages

```
import matplotlib.pyplot as plt
import numpy as np
import time 
from keras.datasets import mnist

(images, labels), (images, labels) = mnist.load_data()
num = len(images)
start = time.time()

average = np.zeros((10, 28, 28))
frequency = np.zeros((10, 1, 1))

for i in np.arange(num):
  label = labels[i]
  average[label] += images[i]
  frequency += 1

average /= frequency
end = time.time()

def display_average():
  plt.figure(figsize=(16, 16))

  for label in np.arange(10):
    plt.subplot(4, 4, 1 + label)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(average[label])

  plt.show()

display_average()
```
