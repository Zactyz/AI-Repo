##First, create the 3D function plot of f1 with p1 minimum

```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x ** 2 + y ** 2

x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

point=ax.scatter(0, 0, 2, color='green')
ax.view_init(60, 35)
```

###Next, the density plot of f1 with p1 minimum

```
x = np.linspace(-5, 8, 50)
y = np.linspace(-5, 8, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();

plt.plot(0, 0, 'o', color='green');
```

###Followed by the contour plot of f1 with p1 minimum

```
x = np.linspace(-5, 8, 50)
y = np.linspace(-5, 8, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[-5, 8, -5, 8], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();

plt.plot(0, 0, 'o', color='green');
```


##Second create the 3D function plot of f2 with p2 maximum

```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return np.cos(x) + np.cos(y)

x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

point=ax.scatter(0, 0, 2, color='green')
ax.view_init(60, 35)
```

###Next, the density plot of f2 with p2 maximum

```
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();

plt.plot(3.1, 3.1, 'o', color='green');
```

###Followed by the contour plot of f2 with p2 maximum

```
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();

plt.plot(3.1, 3.1, 'o', color='green');
```

##Thirdly, create the 3D function plot of f3 with specialcase point p3

```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x ** 3 - y

x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

point=ax.scatter(0, 0, 2, color='green')
```
